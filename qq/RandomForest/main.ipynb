### Случайный лес


Энтропия - мера случайности величины.

Формула (для задачи классификации):
$$
E = -\sum_{i=0}^k p_i \cdot \log_n p_i
$$

Где n - кол-во классов объектов в наших данных.
Где: k - кол-во групп данных, на которые мы делим данные, зачастую = n, НО вычесть 1, так как отсчёт с нуля.
Где $p_i$ - вероятность получение несоответствующей группе величины в наших данных.

Чтобы получить общую энтропию нескольких множеств групп мы используем формулу:  
$$E_1 \cdot x_1/n + E_2 \cdot x_2/n + ...$$

Где $E_i$ и $x_i$ - энтропия и кол-во элементов в множестве групп соответственно, а n - общее кол-во элементов во всех данных (нескольких множествах).


Чтобы не допускать переобучения деревьев им ограничивают глубину, чаще всего 3.
