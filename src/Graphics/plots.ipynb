* Обычный график


```python
from math import e, sin


def func(x):
    return x * sin(x)


def func_sigmoid(x, k):
    return 1 / (e ** (-k * (x - 0.1)))


X = list(range(-100, 100))
Y = list(map(func, X))


plt.plot(
    X,
    Y,
    "b-",
    marker="",
    label="x * sin(x)",
)

plt.plot(
    range(-100, 100),
    [func_sigmoid(x / 100, 5) for x in range(-100, 100)],
    "g:",
    marker="",
    label="sigmoid, k = 5, x0 = 0.1, f(x / 10)",
)

plt.xlabel("x")
plt.ylabel("y")
plt.title("y = x * sin(x) & sigmoid")

plt.legend(loc=8)

# Сохранение диаграммы.
# Сохранять надо до .show(), иначе данные отчистяться,
# и получится белый лист.
plt.savefig(fname="test.svg")

plt.show()
```

* Столбчатые диаграммы


```python
mentions = [500, 505]
years = [2017, 2018]

plt.bar(
    years,
    mentions,
    edgecolor="black",
)

# Дурной тон, так делать не следует.
# Шкала по y всегда должна начинаться от 0.
# Иначе вводит в заблуждение, как в данном случае.
# P.S. Разница всего в 5 упоминаний.
plt.axis([2016.5, 2018.5, 499, 506])

plt.xticks(years)
plt.xlabel("Год")
plt.ylabel("Кол-во упоминаний")
plt.title("Упоминание ML в зависимости от года")

plt.show()
```

* Красивые рисунки


```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = range(len(variance))


plt.plot(xs, variance, ls="solid", color="green", label="Дисперсия")
plt.plot(xs, bias_squared, "r-.", label="Квадрат смещения")
plt.plot(xs, total_error, "b:", label="Суммарная ошибка")

plt.legend(loc=9)
plt.xlabel("Сложность модели")
plt.title("Компромисс между смещением и дисперсией")

plt.savefig(fname="test(2).svg")

plt.show()
```
